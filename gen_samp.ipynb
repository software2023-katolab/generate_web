```python
import sqlite3
import datetime
import random
import math
```

```python
conn = sqlite3.connect('races.sqlite')
cursor = conn.cursor()
```

```python
cursor.execute('DELETE FROM races')
cursor.execute('DELETE FROM horses')
cursor.execute('DELETE FROM runners')
cursor.execute('DELETE FROM predicts')
```

```python
race_id = 1
race_num = 12
race_name = ['3歳未勝利', 'G1', '加藤杯', '1勝クラス']
locations = ['東京', '大阪', '京都']
```

```python
today = datetime.date.today()
current_weekday = today.weekday()
last_saturday = today - datetime.timedelta(days=current_weekday + 2)
last_saturday = last_saturday.strftime('%Y-%m-%d')
last_sunday = today - datetime.timedelta(days=current_weekday + 1)
last_sunday = last_sunday.strftime('%Y-%m-%d')
this_sunday = today + datetime.timedelta(days=6 - current_weekday)
this_sunday = this_sunday.strftime('%Y-%m-%d')
this_saturday = today + datetime.timedelta(days=5 - current_weekday)
this_saturday = this_saturday.strftime('%Y-%m-%d')

dates = [last_saturday, last_sunday, this_saturday]#, this_sunday, ]
```

```python
for date in dates:
    for loca in locations:
        for number in range(race_num):
            number = number+1
            name = random.choice(race_name)
            data = (race_id, date, name, number, loca)
            cursor.execute('INSERT INTO races (id, date, name, number, location) \
                               VALUES (?, ?, ?, ? ,?)', data)
            race_id = race_id + 1
          
conn.commit()  
```

```python
horse_id = 1
horse_name = ['スペシャルウィーク', 'サイレンススズカ', 'トウカイテイオー', 'ゴールドシップ', 'ダイワスカーレット', 'メジロマックイーン'\
             , 'ナリタブライアン', 'マンハッタンカフェ', 'タマモクロス', 'セイウンスカイ', 'グラスワンダー', 'タイトルホルダー']
gender = ['牡', '牝']
```

```python
for horse in horse_name:
    gen = random.choice(gender)
    data = (horse_id, horse, random.randint(1,10), gen)
    cursor.execute('INSERT INTO horses (id, name, age, gender) \
                               VALUES (?, ?, ?, ?)', data)
    horse_id = horse_id+1

conn.commit()
```
